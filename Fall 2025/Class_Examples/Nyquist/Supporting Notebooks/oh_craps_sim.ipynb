# The no pass line in craps

## Winning and Losing
On the come-out roll:
- You win if the shooter rolls a 7 or 11
- You lose if the shooter rolls a 2, 3, or 12 (called "craps")
- If any other number is rolled (4, 5, 6, 8, 9, or 10), that number becomes the "point"
  
If a point is established:
- The shooter continues rolling.
- You win if the point is rolled again before a 7
- You lose if a 7 is rolled before the point (known as "sevening out")

```python
import numpy as np
from datascience import *
```

```python
die = np.arange(1, 7)
```

```python
def roll_dice():
    "Roll two dice"
    result = np.random.choice(die) + np.random.choice(die)
    print(result)
    return result

def craps_round():
    """ Simulate one round of craps """
    roll = roll_dice()
    if roll == 7 or roll == 11:
        print("Natural")
        return 1
    elif roll == 2 or roll == 3 or roll == 12:
        print("Crapped out")
        return 0
    else:
        point = roll
        print("Point is ", point)
    while True:
        roll = roll_dice()
        if roll == point:
            print("Made point")
            return 1
        if roll == 7:
            print("Sevened out")
            return 0  
```

```python
craps_round()
```

```python
def roll_dice():
    "Roll two dice"
    result = np.random.choice(die) + np.random.choice(die)
    return result


def craps_round():
    """ Simulate one round of craps """
    roll = roll_dice()
    if roll == 7 or roll == 11:
        return 1
    elif roll == 2 or roll == 3 or roll == 12:
        return 0
    else:
        point = roll
    while True:
        roll = roll_dice()
        if roll == point:
            return 1
        if roll == 7:
            return 0  
```

```python
craps_round()
```

```python
from tqdm.notebook import tqdm

num_simulations = 1_000_000
win_count = 0
for i in tqdm(np.arange(num_simulations)):
    win_count = win_count + craps_round()

winning_percentage = 100 * win_count / num_simulations
print("The odds of winning at the no-pass line is: ", winning_percentage)
```
