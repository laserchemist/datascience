```python
import numpy as np
from datascience import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers))/np.std(any_numbers) 
    
def correlation(t, label_x, label_y):
    return np.mean(standard_units(t.column(label_x))*standard_units(t.column(label_y)))

def slope(t, label_x, label_y):
    r = correlation(t, label_x, label_y)
    return r*np.std(t.column(label_y))/np.std(t.column(label_x))

def intercept(t, label_x, label_y):
    return np.mean(t.column(label_y)) - slope(t, label_x, label_y)*np.mean(t.column(label_x))

def fit_line(tbl, label_x, label_y):
    # Your code may need more than 1 line below here.
    slope_compute = slope(tbl, label_x, label_y)
    intercept_compute = intercept(tbl, label_x, label_y)
    return make_array(slope_compute, intercept_compute)

def confidence_interval_95(values):
    "Returns an array with the 2.5th and 97.5th percentile of values"
    return np.array([np.percentile(values,2.5), np.percentile(values,97.5)])
```
