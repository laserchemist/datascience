```python
plt.hist(np.random.choice(np.arange(1,101,1),20))
```

```python

tstats = np.array([])
num_observation=500
nums = np.random.choice(np.arange(1,101,1),20)
for i in np.arange(num_observation):    
    side = np.random.choice(['W','D'],20)
    sim = Table().with_columns('Number',nums,'Location',side)
    avgdoor=sim.group('Location',np.average).column('Number average').item(0)
    avgwindow=sim.group('Location',np.average).column('Number average').item(1)
    test_statistic = avgdoor-avgwindow
    #print(test_statistic)
    tstats = np.append(tstats,test_statistic)

# Outside loop
plt.hist(tstats, bins=np.arange(-100,100,5))
```
