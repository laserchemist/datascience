```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('fivethirtyeight')
```

```python
def null_sum(chances=173):
    oneoutcome = np.random.choice([1,-1],chances)
    return np.sum(oneoutcome)
```

```python
null_sum(10)
```

```python
sample = []
for i in np.arange(15000):
    sample.append(null_sum(20))
sample = np.array(sample)
```
