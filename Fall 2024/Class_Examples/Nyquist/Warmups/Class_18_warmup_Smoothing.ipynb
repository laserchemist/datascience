# Smoothing Data

In Lab06 you looked at DNA sequences using different sample sizes and observed that it was easier to see the GC-rich areas with a larger sample size because it smoothed out some of the jitter in the data. Smoothing comes up frequenctly in data analysis and there are many way to accomplish it.

To demonstrate, let's first create some noisy data.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 10, .02)
y = 10 * np.sin(x)
n = len(x)

# Create a random number generator
rng = np.random.default_rng()

# Create random normal noise with mean zero, sigma=3
noise = rng.normal(loc=0, scale=3, size=n)

# Add the noise to our data
yn = y + noise
```

```python
plt.plot(x, yn);
```

```python
def moving_average(arr, window_size):
    
    # Prepare an array to store the moving averages
    moving_averages = np.empty(len(arr) - window_size + 1)
    
    # Calculate the moving average using a loop
    for i in range(len(moving_averages)):
        moving_averages[i] = np.mean(arr[i:i + window_size])
    
    return moving_averages
```

```python
# Number of points to average
window = 10
```

```python
y_10 = moving_average(yn, window)

len(y_10)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, yn)
axes[0].set_title("Unfiltered")
axes[1].plot(x[0:len(y_10)], y_10)
axes[1].set_title(f"Filtered window size {window}");
```
